```python
df=pd.read_csv('../input/Fuel Consumption')
```

```python
df.head(5)
```

```python
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
print(df.shape)
for i in range(df.shape[0]):
    if(df.loc[i].isnull().sum()>5):
        df.drop([i],inplace=True)
print(df.shape)
```

```python
df.isnull().any()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.drop(columns=['MODEL', 'MAKE', 'MODEL.1'],inplace=True)
```

```python
df
```

```python
CO2_raw = df['CO2 EMISSIONS ']
features_raw = df.drop('CO2 EMISSIONS ', axis = 1)
features_raw = features_raw.drop('VEHICLE CLASS', axis = 1)
```

```python
features_raw
```

```python
features_raw.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['ENGINE SIZE', 'CYLINDERS', 'FUEL CONSUMPTION', 'Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11']
features_raw[numerical].astype(float)
features_raw[numerical]=scaler.fit_transform(features_raw[numerical])
display(features_raw.head(5))
```

```python
features_final = pd.get_dummies(features_raw)
```

```python
features_final.columns
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_final, 
                                                    CO2_raw, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=50, random_state=0,n_estimators=100)
regr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import explained_variance_score
y_pred=regr.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
explained_variance_score(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
